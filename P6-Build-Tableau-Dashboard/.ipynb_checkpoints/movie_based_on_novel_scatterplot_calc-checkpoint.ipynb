Find proportion of movies that have `vote_average` > average `vote_average` and `profit_adj` > average profit_adj for each category - `based_on_novel` True and False. I call this quadrant magic quadrant.

```python
magic_quadrant = (movies_cleaned["based_on_novel"][(movies_cleaned["vote_average"] > 5.947) & 
        (movies_cleaned["profit_adj"] > 41937684)].value_counts())/(movies_cleaned["based_on_novel"].value_counts())
magic_quadrant
```
